### 1.1 Data Cleaning
Remove the records with missing coordinates and those that are duplicates and extract temporal features from Report Date.

```python
print(f"Records before cleaning: {len(df_uf)}")
df_uf_clean = df_uf.dropna(subset=['Latitude', 'Longitude']).copy()
print(f"Records after removing missing coords: {len(df_uf_clean)}")
print(f"Dropped: {len(df_uf) - len(df_uf_clean)} records\n")

# Note: UF Crime logs contain multiple rows per incident tracking case status updates.
# We keep only the first record for each unique report number.
print(f"Records before clean: {len(df_uf_clean)}")
df_uf_clean['Report Date'] = pd.to_datetime(df_uf_clean['Report Date'])
df_uf_clean = df_uf_clean.sort_values('Report Date')
df_uf_clean = df_uf_clean.drop_duplicates(subset='Report Number', keep='first')
print(f"Records after deduplication: {len(df_uf_clean)}")
print(f"Removed: {934 - len(df_uf_clean)} duplicate status updates\n")

df_uf_clean['Year'] = df_uf_clean['Report Date'].dt.year
df_uf_clean['Month'] = df_uf_clean['Report Date'].dt.month
df_uf_clean['Month_Name'] = df_uf_clean['Report Date'].dt.month_name()
df_uf_clean['Day_of_Week'] = df_uf_clean['Report Date'].dt.day_name()

print("Data Cleaning Complete")
print(f"Final dataset: {len(df_uf_clean)} unique bike theft incidents")
print(f"Date range: {df_uf_clean['Report Date'].min()} to {df_uf_clean['Report Date'].max()}")
```

### 2.1 Database Schema Design
Include the most relevant columns for our analysis, excluding uneccesary columns.

```python
# Excluding Crime Start, Crime End, and Log Date

create_table_sql = '''
CREATE TABLE IF NOT EXISTS bike_thefts (
    report_number INTEGER PRIMARY KEY,
    report_date TEXT NOT NULL,
    agency_name TEXT,
    crime_type TEXT,
    disposition TEXT,
    building TEXT,
    address TEXT NOT NULL,
    latitude REAL NOT NULL,
    longitude REAL NOT NULL,
    year INTEGER,
    month INTEGER,
    month_name TEXT,
    day_of_week TEXT
);
'''

print("Database Schema:")
print(create_table_sql)
print("\nThis table will store all 581 bike theft records with:")
print("  - Temporal info: report_date, year, month, day_of_week")
print("  - Location info: building, address, latitude, longitude")
print("  - Case info: disposition")
```

## 3. Temporal Pattern Analysis

Identify when bike thefts are most likely to occur by analyzing monthly and yearly patterns.

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

conn = sqlite3.connect('database/bike_thefts.db')
df = pd.read_sql_query("SELECT * FROM bike_thefts", conn)
conn.close()

df['report_date'] = pd.to_datetime(df['report_date'])

print(f"Loaded data:")
print(f"Total theft incidents: {len(df)}")
print(f"Date range: {df['report_date'].min()} to {df['report_date'].max()}")
```

### 3.1 Monthly Theft Patterns
Analyzing which months experience the highest theft rates across all years.

```python
monthly_thefts = df.groupby('month').size().reset_index(name='theft_count')

monthly_thefts = monthly_thefts.sort_values('month')

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_thefts['month_abbr'] = monthly_thefts['month'].apply(lambda x: month_names[x-1])

print("Thefts by Month:")
print(monthly_thefts[['month_abbr', 'theft_count']].to_string(index=False))
print()

peak_month = monthly_thefts.loc[monthly_thefts['theft_count'].idxmax()]
lowest_month = monthly_thefts.loc[monthly_thefts['theft_count'].idxmin()]

print(f"Peak month: {peak_month['month_abbr']} with {peak_month['theft_count']} thefts")
print(f"Lowest month: {lowest_month['month_abbr']} with {lowest_month['theft_count']} thefts")
print(f"Average thefts per month: {monthly_thefts['theft_count'].mean():.1f}")
```

### 3.2 Yearly Theft Trends
Analyzing how bike theft rates have changed from 2016 to 2022.

```python
yearly_thefts = df.groupby('year').size().reset_index(name='theft_count')

print("Thefts by Year:")
print(yearly_thefts.to_string(index=False))
print()

yearly_thefts['yoy_change'] = yearly_thefts['theft_count'].diff()

print("Year by Year Change:")
for idx, row in yearly_thefts.iterrows():
    if pd.notna(row['yoy_change']):
        direction = "up" if row['yoy_change'] > 0 else "down"
        print(f"{int(row['year'])}: {int(row['theft_count'])} thefts "
              f"({direction} {abs(row['yoy_change']):.0f})")
    else:
        print(f"{int(row['year'])}: {int(row['theft_count'])} thefts (baseline)")

print()

peak_year = yearly_thefts.loc[yearly_thefts['theft_count'].idxmax()]
print(f"Peak year: {int(peak_year['year'])} with {int(peak_year['theft_count'])} thefts")
print(f"Note: 2016 only has 1 record")
print(f"Note: 2022 only has 10 records")
```

### 3.3 Monthly Patterns by Year
Examining whether peak months are consistent across different years.

```python
monthly_by_year = df.groupby(['year', 'month']).size().reset_index(name='theft_count')

pivot = monthly_by_year.pivot(index='month', columns='year', values='theft_count').fillna(0).astype(int)

pivot.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

print("Thefts by Month/Year:")
print(pivot)
print()

print("Peak Month Per Year (excluding 2016/2022):")
for year in [2017, 2018, 2019, 2020, 2021]:
    year_data = monthly_by_year[monthly_by_year['year'] == year]
    peak = year_data.loc[year_data['theft_count'].idxmax()]
    month_name = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'][int(peak['month'])-1]
    print(f"{int(year)}: {month_name} ({int(peak['theft_count'])} thefts)")
```

### 3.4 Temporal Pattern Summary
Key statistics and findings from temporal analysis.

```python
print("Temporal Pattern Analysis: Summary of Statistics")
print()

print("Overall Statistics:")
print(f"Total thefts analyzed: {len(df)}")
print(f"Time period: Aug 2016 - Mar 2022 ({(df['report_date'].max() - df['report_date'].min()).days} days)")
print(f"Average thefts per month: {len(df) / 66:.1f}")
print(f"Average thefts per year: {len(df) / 5.6:.1f}")
print()

print("Monthly Patterns:")
print(f"Highest month overall: September (88 thefts)")
print(f"Lowest month overall: May (25 thefts)")
print(f"Fall semester (Aug-Nov): {df[df['month'].isin([8,9,10,11])]['month'].count()} thefts")
print(f"Spring semester (Jan-Apr): {df[df['month'].isin([1,2,3,4])]['month'].count()} thefts")
print(f"Summer (May-Jul): {df[df['month'].isin([5,6,7])]['month'].count()} thefts")
print()

print("Yearly Trends:")
print(f"Peak year: 2018 (157 thefts)")
print(f"Notable decline of thefts since 2018 onwards")
print(f"Most consistent peak months: September/October")
print()

print("Interpretations:")
print("- Bike thefts strongly correlate with academic calendar as thefts often occur towards the start of the semesters")
print("- Fall semester is highest risk, which can be attributed to new students coming in with little experience not locking up their bikes properly or being in a rush to a new location for class and not locking up correctly.")
print("- Summer months show lowest theft rates due to fewer students on campus")
print("- Overall decline from 2018 peak may indicate improved security")
```

## 4. Spatial Pattern Analysis
Identify geographic patterns and determine optimal clustering strategy for identifying risk zones.

```python
sns.set_style("whitegrid")

conn = sqlite3.connect('database/bike_thefts.db')
df = pd.read_sql_query("SELECT * FROM bike_thefts", conn)
conn.close()

print("Data Loaded for Spatial Analysis")
print(f"Total theft incidents: {len(df)}")
print(f"Records with coordinates: {df[['latitude', 'longitude']].notna().all(axis=1).sum()}")
```

### 4.1 Top Theft Locations
Identifying buildings and addresses with highest theft frequency.

```python
print("Top 10 Highest Theft Prone Buildings:")
building_counts = df[df['building'].notna()].groupby('building').size().reset_index(name='theft_count')
building_counts = building_counts.sort_values('theft_count', ascending=False).head(10)
print(building_counts.to_string(index=False))
print(f"\nNote: {df['building'].isna().sum()} records have no building name")
print()

print("Top 10 Highest Theft Prone Addresses:")
address_counts = df.groupby('address').size().reset_index(name='theft_count')
address_counts = address_counts.sort_values('theft_count', ascending=False).head(10)
print(address_counts.to_string(index=False))
print()

print("Geographic Spread")
print(f"Unique buildings: {df['building'].nunique()}")
print(f"Unique addresses: {df['address'].nunique()}")
print(f"Latitude range: {df['latitude'].min():.6f} to {df['latitude'].max():.6f}")
print(f"Longitude range: {df['longitude'].min():.6f} to {df['longitude'].max():.6f}")
```

### 4.2 Prepare Data for Clustering
Extract and examine coordinate data for K-means clustering analysis.

```python
X = df[['latitude', 'longitude']].values

print("Clustering Data Prep:")
print(f"Feature matrix shape: {X.shape}")
print(f"Total data points for clustering: {X.shape[0]}")
print()

print("Coordinate Stats:")
print(f"Latitude: Mean = {X[:, 0].mean():.6f}, Std = {X[:, 0].std():.6f}")
print(f"Longitude: Mean = {X[:, 1].mean():.6f}, Std = {X[:, 1].std():.6f}")
print()

lat_range = X[:, 0].max() - X[:, 0].min()
lon_range = X[:, 1].max() - X[:, 1].min()
print(f"Latitude range: {lat_range:.6f} degrees (~{lat_range * 69:.2f} miles)")
print(f"Longitude range: {lon_range:.6f} degrees (~{lon_range * 54:.2f} miles)")
print()

print("Coordinate data ready for clustering!")
```

### 4.3 Elbow Method for Optimal K
Testing different numbers of clusters to determine the optimal number of risk zones.

```python
K_range = range(2, 11)  # Testing K from 2 to 10
inertias = []

print("Testing Different K values:")

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
    print(f"K = {k}: Inertia = {kmeans.inertia_:.6f}")

plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.title('Elbow Method for Optimal K', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(K_range)

for i, (k, inertia) in enumerate(zip(K_range, inertias)):
    plt.annotate(f'{inertia:.4f}', 
                xy=(k, inertia), 
                textcoords='offset points',
                xytext=(0, 10), 
                ha='center',
                fontsize=9)

plt.tight_layout()
plt.show()

print("\nElbow curve plotted!")
```

```python
print("Inertia Reduction Analysis")
print(f"{'K':<5} {'Inertia':<12} {'Reduction':<12} {'% Improvement':<15}")

for i, (k, inertia) in enumerate(zip(K_range, inertias)):
    if i == 0:
        print(f"{k:<5} {inertia:<12.6f} {'—':<12} {'—':<15}")
    else:
        reduction = inertias[i-1] - inertia
        pct_improvement = (reduction / inertias[i-1]) * 100
        print(f"{k:<5} {inertia:<12.6f} {reduction:<12.6f} {pct_improvement:<15.2f}%")

print("\nInterpretation:")
print("The 'elbow' is where % improvement drops significantly.")
print("After the elbow, adding more clusters doesn't improve the spread significantly.")
```

```python
print("Elbow Analysis Conclusions:")
print("\nBased on the inertia reduction analysis:")
print("K=2-5: Strong improvements")
print("K=6-7: Moderate improvements")
print("K=8+: Weak improvements")
print()
print("Visual Conclusion: K = 5 for elbow as graph really starts to transition from steep to flat here.")
```

## 5. K-means Clustering & Risk Zone Identification
Create 5 geographic risk zones using K-means clustering and validate cluster quality.

```python
conn = sqlite3.connect('database/bike_thefts.db')
df = pd.read_sql_query("SELECT * FROM bike_thefts", conn)
conn.close()

X = df[['latitude', 'longitude']].values

print("K-means Clustering with K=5")
print(f"Data points: {len(X)}")
print(f"Features: Latitude, Longitude")
```

### 5.1 Apply K-means with K=5

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X)

df['risk_zone'] = df['cluster'] + 1

cluster_centers = kmeans.cluster_centers_

print("Clustering Complete")
print(f"Number of clusters: 5")
print(f"Cluster labels assigned: {df['cluster'].nunique()}")
print()

print("Cluster Centers (Geographic Centroids):")
for i in range(5):
    center_lat, center_lon = cluster_centers[i]
    print(f"Cluster {i+1}: Lat={center_lat:.6f}, Lon={center_lon:.6f}")
print()

cluster_sizes = df['cluster'].value_counts().sort_index()
print("Number of Thefts per Zone:")
for i in range(5):
    count = cluster_sizes[i]
    percentage = (count / len(df)) * 100
    print(f"Cluster {i+1}: {count} thefts")
```

### 5.2 Silhouette Analysis
Silhouette score measures how cohesive and efficient the clusters are. The closer the score is to 1, the stronger and better separated the clusters are.

```python
silhouette_avg = silhouette_score(X, df['cluster'])

print("Silhouette Analysis")
print(f"Silhouette Score: {silhouette_avg:.3f}")
print()

if silhouette_avg >= 0.7:
    quality = "Excellent, clusters are strong and well structured"
elif silhouette_avg >= 0.5:
    quality = "Solid, clusters are reasonably structured"
elif silhouette_avg >= 0.25:
    quality = "Decent, cluster structure is there but weak"
else:
    quality = "Poor, cluster structure is insignificant"

print(f"Interpretation: {quality}")
print()

sample_silhouette_values = silhouette_samples(X, df['cluster'])

print("Silhouette Score by Cluster:")
for cluster_id in range(5):
    cluster_silhouette_values = sample_silhouette_values[df['cluster'] == cluster_id]
    avg_cluster_score = cluster_silhouette_values.mean()
    print(f"Cluster {cluster_id+1}: {avg_cluster_score:.3f} (n={cluster_sizes[cluster_id]})")
```

### 5.3 Cluster Interpretation
Analyzing which buildings and locations fall into each cluster.

```python
print("Risk Zone Analysis:")

for cluster_id in range(5):
    cluster_data = df[df['cluster'] == cluster_id]
    print(f"Cluster {cluster_id+1} - Risk Zone {cluster_id+1}")
    print(f"Total thefts: {len(cluster_data)}")
    print(f"Center coordinates: Lat={cluster_centers[cluster_id][0]:.6f}, Lon={cluster_centers[cluster_id][1]:.6f}")
    print(f"Silhouette score: {sample_silhouette_values[df['cluster'] == cluster_id].mean():.3f}")
    print()
    
    buildings_in_cluster = cluster_data[cluster_data['building'].notna()]['building'].value_counts().head(5)
    if len(buildings_in_cluster) > 0:
        print("Top 5 buildings:")
        for building, count in buildings_in_cluster.items():
            print(f"  • {building}: {count} thefts")
    else:
        print("No building names recorded for this cluster")
    print()
    
    addresses_in_cluster = cluster_data['address'].value_counts().head(5)
    print("Top 5 addresses:")
    for address, count in addresses_in_cluster.items():
        print(f"  • {address}: {count} thefts")
    print()
```

```python
risk_labels = {
    0: "Southwest Campus Zone",
    1: "North Campus Zone",
    2: "East Campus Residential Zone",
    3: "Central Campus Zone",
    4: "Tanglewood Zone"
}

df['risk_zone_name'] = df['risk_zone'].map(risk_labels)

risk_levels = {
    0: "Low to Moderate",
    1: "Moderate to High",
    2: "Very High",
    3: "Moderate to High",
    4: "Very Low"
}

df['risk_level'] = df['risk_zone'].map(risk_levels)

print("Risk Zone Summary:")
for cluster_id in range(5):
    print(f"\nCluster {cluster_id+1}: {risk_labels[cluster_id]}")
    print(f"Risk Level: {risk_levels[cluster_id]}")
    print(f"Thefts: {cluster_sizes[cluster_id]}")
    print(f"Silhouette: {sample_silhouette_values[df['cluster'] == cluster_id].mean():.3f}")

print()
print("Risk zones identified and labeled!")
```

**Insights/Interpretations:**

Zone 3 is the highest risk zone with 226 thefts, which is comprised of the most popular and major dorms such as Broward, Beaty, and Jennings.
Silhouette score of 0.475 validates our clustering to be decent.
Clear spatial separation between zones, especially Zone 5

**Why this is happening:**

Dorms are the most jam packed with students on campus who leave their bikes right outside over night. This zone is where the most popular and largest dorms on campus are. Additionally, some of these dorms are only allowed for freshmen to take. As mentioned earlier, freshmen are new to campus and new to protecting their bikes, so it may take some getting used to from them to ensure their bikes are properly locked and protected. The start of their first semester (typically fall) makes their bikes in those areas especially vulnerable. Until they figure out how to secure their bike for sure, criminals can take advantage of that fact.

**What to do about it:**

Increase bike rack security and lighting at night at the dorms.
Launch additional patrols around these dorms during Fall semester where it happens more frequently.

```python
silhouette_avg = silhouette_score(X, df['cluster'])
sample_silhouette_values = silhouette_samples(X, df['cluster'])

fig, ax = plt.subplots(figsize=(10, 8))

y_lower = 10
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']

for cluster_id in range(5):
    cluster_silhouette_values = sample_silhouette_values[df['cluster'] == cluster_id]
    cluster_silhouette_values.sort()
    
    size_cluster = cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster
    
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, cluster_silhouette_values,
                     facecolor=colors[cluster_id], edgecolor=colors[cluster_id], alpha=0.7)
    
    ax.text(-0.05, y_lower + 0.5 * size_cluster, f'Cluster {cluster_id+1}')
    
    y_lower = y_upper + 10

ax.axvline(x=silhouette_avg, color="red", linestyle="--", linewidth=2, 
           label=f'Average Score: {silhouette_avg:.3f}')

ax.set_xlabel('Silhouette Score', fontsize=13, fontweight='bold')
ax.set_ylabel('Cluster Score', fontsize=13, fontweight='bold')
ax.set_title('Silhouette Plot for K-means Clustering (K=5)', 
             fontsize=15, fontweight='bold')
ax.legend(loc='best', fontsize=11)
ax.set_yticks([])
ax.set_xlim([-0.1, 1])

plt.tight_layout()
plt.savefig('visualizations/silhouette_plot.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved as visualizations/silhouette_plot.png")
```

**Note**: Each sliver in the bars represents a few bike thefts and their individual silhouette scores with their cluster. For example, one sliver of orange goes backwards in cluster 2 indicating it should have been assigned to a different cluster.

```python
K_range = range(2, 11)
inertias = []

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(X)
    inertias.append(kmeans_temp.inertia_)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(K_range, inertias, 'bo-', linewidth=2, markersize=10)

ax.plot(5, inertias[3], 'r*', markersize=20, label='Selected Value')

ax.set_xlabel('Number of Clusters (K)', fontsize=13, fontweight='bold')
ax.set_ylabel('Inertia', fontsize=13, fontweight='bold')
ax.set_title('Elbow Method for Optimal K', fontsize=15, fontweight='bold')
ax.set_xticks(K_range)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

for i, (k, inertia) in enumerate(zip(K_range, inertias)):
    ax.annotate(f'{inertia:.4f}', 
                xy=(k, inertia), 
                textcoords='offset points',
                xytext=(0, 10), 
                ha='center',
                fontsize=9)

plt.tight_layout()
plt.savefig('visualizations/elbow_curve.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved as visualizations/elbow_curve.png")
```
